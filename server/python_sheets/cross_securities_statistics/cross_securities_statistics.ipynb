В этом листке попробуем сравнить статистику по нескольким символам между собой, чтобы определить тенденции в лидерстве среди наиболее популярных криптовалют.
Для этого выберем 5 самых популярных торговых единиц на текущий момент: BTC, BFX, ETC, ETH, LTC.

Традиционный шаг: импортируем вспомогательную библиотеку - визуализатор:

```python
from pandas import date_range,Series,DataFrame,read_csv, qcut
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Удобнее всего соотносить криптовалюты, используя статистику по торговле с долларом. Таким образом можно построить совместную визуализацию объемов, средних цен, динамики роста и многого другого. Чтож, начнем со считывания данных:

```python
btc_usd_data = read_csv('../../data/vwapHourlyBTCUSD.csv', header=0)
bfx_usd_data = read_csv('../../data/vwapHourlyBFXUSD.csv', header=0)
etc_usd_data = read_csv('../../data/vwapHourlyETCUSD.csv', header=0)
eth_usd_data = read_csv('../../data/vwapHourlyETHUSD.csv', header=0)
ltc_usd_data = read_csv('../../data/vwapHourlyLTCUSD.csv', header=0)
```

Начнем с того, что посчитаем для каждого инструмента суммарный проторгованный объем.

```python
btc_usd_turnover = np.sum(btc_usd_data['Volume'])
bfx_usd_turnover = np.sum(bfx_usd_data['Volume'])
etc_usd_turnover = np.sum(etc_usd_data['Volume'])
eth_usd_turnover = np.sum(eth_usd_data['Volume'])
ltc_usd_turnover = np.sum(ltc_usd_data['Volume'])
```

Отобразим эти значения на круговой диаграмме:

```python
plt.figure(figsize=(7,5))
x = [btc_usd_turnover, bfx_usd_turnover, etc_usd_turnover, eth_usd_turnover, ltc_usd_turnover]
labels = ['LTC', 'BFX', 'BTC', 'ETC', 'ETH']
explode = [0, 0, 0.05, 0, 0]

plt.pie(x, labels = labels, explode = explode, autopct = '%1.1f%%', shadow=True);
plt.show()
```

На этом графике отчетливо видно, что на данный момент биткоин занимает больше трети всего проторгованного объема. Это объясняет его привлекательность для инвесторов и высокую ликвидность.

Второе место стабильно занимает так называемый "Эфир" (ETH). Многие аналитики предсказывают его скорое первенство, что, на самом деле, вполне возможно: его объемы отстают от биткойна всего на 14%.

Посмотрим теперь на средние цены валют. По выдвинутому мной предположению о следствии высокой ликвидности от объема, должна прослеживаться корреляция с предыдущей круговой диаграммой.

```python
Посчитаем средние цены на текущий момент:
```

```python
btc_middle_price = btc_usd_data["Vwap (BTCUSD)"][0]
bfx_middle_price = bfx_usd_data["Vwap (BFXUSD)"][0]
etc_middle_price = etc_usd_data["Vwap (ETCUSD)"][0]
eth_middle_price = eth_usd_data["Vwap (ETHUSD)"][0]
ltc_middle_price = ltc_usd_data["Vwap (LTCUSD)"][0]
```

Вполне очевидно, что в качестве визуализирующего инструмента в данном случае нужно взять столбчатую диаграмму.

```python
x = ['BTC','BFX','ETC ','ETH' ,'LTC']
z1 = [btc_middle_price, bfx_middle_price, etc_middle_price, eth_middle_price, ltc_middle_price]

# bar()
fig = plt.figure()
plt.bar(x, z1)
plt.title('Middle prices')
plt.grid(True)   # линии вспомогательной сетки
plt.show()
```

Ожидаемо, сохранились первые два места. Однако можно увидеть, что дальше не все так однозначно: Litecoin имеет цену бОльшуу, чем ClassicEtherium, хотя на предыдущей диаграмме отношение было обратным. Этот факт легко объясняется небольшим возрастом валюты Litecoin: первая транзацкия по ней была проведена всего полгода назад, зато за это время она быстро приобрела популярность и обогнала ETC.

Итак, как мы можем видеть, сравнительная статистика довольно продуктивна: находятся новые закономерности, интересные факты и т. п.
